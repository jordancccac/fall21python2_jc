```python
#determine rate of redacted call descriptions across ems dispatches 
#has rate of redacted changed? how so? 
#dictonary keys = year of dispatch, value=# of records whoâ€™s description is redacted)

'''This function reads the file and returns it to be used with other functions; 'with' 
is not being used to open the file for reading becuase 'with' automatically closes the file. It uses dict reader
to return a dictionary'''

#import csv module 
import csv

def read_file():
    
    #open file for reading globally  
    ems_csv=open('ac_ems_300records.csv','r')
    
    #use dict reader to put records into a dictionary 
    reader=csv.DictReader(ems_csv)

    return reader 


```

```python

def test_for_redact():
    
    '''this function will iterate over each row, testing it for a redacted description, and adding those items to a list'''

    #pass the readfile to this function: 
    reader=read_file()
    
    #create empty list to hold redacted calls for each year
    totalredactcalls=[]
      
     #add redacted calls by year to a list 
    for record in reader:
        
        if record['description_short']=="Removed":
            totalredactcalls.append(record['call_year'])
            
    #return the list 
    return totalredactcalls 

    
```

```python
from collections import Counter

def count_redact():
    '''This function counts the yearly redacted calls
    Input: this function takes in the above list of redacted calls and counts it into a dictionary.
    '''
    #pass the list of redacted calls to this function 
    totalredactcalls=test_for_redact()
    
    #pass the list to the dictionary counter 
    trc_counter=Counter(totalredactcalls)


    return trc_counter 
```

```python
from collections import Counter 

def total_yearly_calls():

    
    '''This function totals yearly calls and displays them to the console. It takes in the redacted calls 
    dictionary from the count_redact function, so it can be combined with the total calls dictionary to create the complete 
    dictionary of both numbers with the year.'''
    
    #pass redacted counter function to this function to get the redacted call list dictionary to this function 
    trc_counter=count_redact()

    
    #pass the read file to this function
    reader=read_file()
    
    #create empty list to hold yearly totals 
    total_ann_calls=[]
    
    #loop through the records 
    for record in reader:
        #add calls to the list 
        total_ann_calls.append(record['call_year'])
          
    
    #pass list to dictionary counter function 
    total_calls_counter=Counter(total_ann_calls)
    
    #use loop to count and display totaly yearly calls 
    for key, value in sorted(total_calls_counter.items()):
        print(key,'had',value,'total EMS calls.')
    print()
        
       #loop through redacted calls and display the counts of each year 
    for key, value in sorted(trc_counter.items()):
        print(key,'had',value,'redacted calls.')
   
        
    #loop through total calls dictionary and add the values from the redacted calls dictionary
    #(combining the 2 dictionaries)
    for key in total_calls_counter:
        total_calls_counter[key]=[total_calls_counter[key]],[trc_counter[key]]
            
    print()
    
    print(total_calls_counter)
    
    print()
    print('Sorted items: ',sorted(total_calls_counter.items()))
        
```

```python
import csv

def write_ID():
    
    '''This function will receive the read file of records from the above function. It will open a new file for writing,
    loop through the readfile, testing for redacted descriptions. Qualifying records will be written to a new csv file.'''
    
    
    #pass in the read file from previous function
    reader=read_file()
    
    #open new file for writing
    with open('redact_ID.csv','w',newline='')as csvID:
        
        #use writer to write onto rows
        writer=csv.writer(csvID)
        
        #write fieldname/header row
        writer.writerow(['call_id'])
        
        #loop through the read file for redacteds and add them to the file by call ID
        
        for record in reader: 
            if record['description_short'] =="Removed":
                writer.writerow({record['_id']})
        print()
        #print confirmation message
        print('Redacted IDs written to file.')
        
        
        
        
```

```python
'''This cell takes the initiating action (i.e. the guts of a main function), by calling the functions here'''


total_yearly_calls()

write_ID()
```
